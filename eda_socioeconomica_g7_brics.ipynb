##Análise de dados comparativa entre países do G7 com países do BRICS

Os dados foram retirados dos portais:

  O World Inequality Database visa fornecer acesso aberto e conveniente ao mais extenso banco de dados disponível sobre a evolução histórica da distribuição mundial de renda e riqueza, tanto dentro dos países quanto entre países.

* Gapminder: https://www.gapminder.org/

  A Gapminder é uma fundação sueca independente, sem afiliações políticas, religiosas ou econômicas que identifica equívocos sistemáticos sobre importantes tendências e proporções globais e usa dados confiáveis ​​para desenvolver materiais didáticos fáceis de entender para livrar as pessoas de seus equívocos.

#Tratamento dos Dados

Os países que não pertecem ao G7 e BRICS deverão serem excluídos da análise assim como os dados faltantes. O índice numérico também será excluído e substituído pelos nomes dos países e manteremos apenas o ano mais recente disponível.

**Índice de desenvolvimento humano (HDI):**

```python
#G7
# Seleção dos países
g7hdi = hdi[hdi["country"].isin(["Germany", "Canada", "United States", "France", "Italy", "Japan", "United Kingdom"])]
# Deleta os dados faltantes
g7hdi = g7hdi.dropna(axis=1)
# Transforma o índice nos nomes dos países
g7hdi = g7hdi.set_index("country")
# Seleciona apenas o ano mais recente disponível
g7hdi = g7hdi.iloc[:,-1:]

#Visualiza os dados
display(g7hdi.shape)
display(g7hdi.head())
display(g7hdi.tail())

#BRICS
bricshdi = hdi[hdi["country"].isin(["Brazil", "Russia", "India", "China", "South Africa"])]
bricshdi = bricshdi.dropna(axis=1)
bricshdi = bricshdi.set_index("country")
bricshdi = bricshdi.iloc[:,-1:]

display(bricshdi.shape)
display(bricshdi.head())
display(bricshdi.tail())
```

#Análise dos Dados

Agora que os dados já estão tratados e disponibilizados de forma apropriada, iremos fazer algumas comparações.

**Índice de Desenvolvimento Humano:**

```python
# Plot
g7hdi.plot.bar()
plt.show()
print("A média é:", g7hdi['2018'].mean())
```

```python
# Plot
bricshdi.plot.bar()
plt.show()
print("A média é:", bricshdi['2018'].mean())
```

O IDH leva diversos ítens em sua composição, sendo os principais referentes a saúde, educação e renda.

Países com IDH até 0,499 são considerados de desenvolvimento humano baixo, os com índices entre 0,50 e 0,799 são considerados de desenvolvimento humano médio e, acima de 0,8 é considerado elevado.

A média do IDH do G7 é considerada elavada enquanto do BRICS é média.

Os países do G7 possuem em média um índice de desenvolvimento humano 24% maior do que os países do BRICS.

**Índice de inflação:**

Uma inflação elevada impacta diretamente na alta dos preços de produtos e serviços, sendo sentidos com maior intensidade nas classes mais vulneráveis.

O BRICS possui um índice de inflação 69% maior em relação ao G7.

#Conclusão
